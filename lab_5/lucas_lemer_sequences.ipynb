```python
d = 3

U_seq = [0, 1]
V_seq = [2, d]

for _ in range(2, 50):
    # Find next elements by ops with previous two k
    U_k = U_seq[-2]
    U_k1 = U_seq[-1]
    U_seq.append(d * U_k1 - U_k)

    V_k = V_seq[-2]
    V_k1 = V_seq[-1]
    V_seq.append(d * V_k1 - V_k)

print(f"U: {U_seq}\n")
print(f"V: {V_seq}\n")

print(len(U_seq))
```
